```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
import warnings
warnings.filterwarnings('ignore')
print("Libraries imported successfully!")
```

```python
# Load Dataset
df = pd.read_csv('vehicles.csv', engine='python', on_bad_lines='skip')
print("Shape:", df.shape)
print(df.head())
```

```python
# Check Missing Values
print("Missing Values:")
print(df.isnull().sum())
print("\nTotal missing:", df.isnull().sum().sum())
```

```python
# Initial Statistics
print(df.describe())
```

```python
# Data Types
print("Data Types:")
print(df.dtypes)
print("\nDimensions:", df.shape)
```

```python
# Select relevant columns and fill missing values
df = df[['price','year','odometer','fuel','transmission','condition','manufacturer']].copy()
df['year'].fillna(df['year'].median(), inplace=True)
df['odometer'].fillna(df['odometer'].median(), inplace=True)
df['fuel'].fillna('unknown', inplace=True)
df['transmission'].fillna('unknown', inplace=True)
df['condition'].fillna('good', inplace=True)
df['manufacturer'].fillna('unknown', inplace=True)

# Type conversion
df['year'] = df['year'].astype(int)
print("After cleaning:")
print(df.dtypes)
```

```python
# Normalize numeric columns
scaler = MinMaxScaler()
df[['price_norm','odometer_norm']] = scaler.fit_transform(df[['price','odometer']])
print("Normalized columns added:")
print(df[['price','price_norm','odometer','odometer_norm']].head())
```

```python
# Encode Categorical Variables
le = LabelEncoder()
for col in ['fuel','transmission','condition','manufacturer']:
    df[col+'_enc'] = le.fit_transform(df[col])
    print(f"{col} encoded.")
print("\nFinal DataFrame:")
print(df.head())
```
